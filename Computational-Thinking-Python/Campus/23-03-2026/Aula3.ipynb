```python
#Excercícios 

# 1 - Sistema de login com nivel de acesso. Solicite usuário e senha. Se usario é igual a admin, crie uma estutura de condição alinhada para solicitar senha e se a mesma for '1234' mostre que o usuario terá acesso total. Caso o usuário insira a senha incorreta, mostre senha incorreta. Caso usuário insira usuário incorreto mostre usuário incorreto 

usuario = input ("Digite seu usuario")
senha = int (input ("Digite sua senha "))

if usuario == "admin":
    if senha == 1234:
        print("Acesso total!!")
    else:
        print("Senha incorreta!")

else:
    print("Usuario incorreto")



 
```

```python
#2 - Classificação de idade. Solicite idade, se idade for maior ou igual a 18, crie um estrutura de condição aninhada para verificar se idade é maior ou igual a 60, se for, mostre que é idoso, senão mostre que é adulto. Se idade for maior ou igual a 12, adolescnete, caso contrário, criança.

idade = int(input ("Digite sua idade"))

if idade >= 18:
    if idade >= 60:
        print("Velho")
    else:
        print("adulto")

elif idade >=12:
    print("Adolescente")
else:
    print ("Criança")
```

```python
# 3 - Aprovação com distinção. solicite nota, se nota for maior ou igual a 6, crie condição aninhada para verificar se nota é maior ou igual a 9, se for, aprovada com excelência. Se nota não for maior ou igual a 9, aprovada. Caso contrário, reprovado

nota = int (input ("Digite sua nota"))

if nota >= 6:
    if nota >= 9:
        print("Aprovada com êxito!")
    else:
        print("Aprovada")
else:
    print("Reprovada")
    
```

```python
#4 - Verficação de número. Solicite número e verifique se é maior do que zero, se for, crie uma estrutura de condição aninhada para verificar se este numero é par, se for, print positivo e par. Se não for, positivo é impar. Se número for igual a zero, print zero, caso contrário negativo.

numero = int(input ("Digite o numero"))

if numero >= 0:
    if numero % 2 == 0:
        print ("numero par")
    else:
        print ("numero impar")
else:
    print ("Zero")
```

```python
#5 - Sistema de desconto. Solicite valor e se a pessoa é vip ou não. if valor maior ou igual 200, crie estrutura de condição aninhada para verificar se a pessoa é vip, se for ofereça 20% sobre o valor e mostre o valor a se descontado e o valor final, considerando o desconto. Se não for vip, ofereça o desconto de 10% 

valor = int (input("Digite o valor"))
Vip = input ("Você faz pate dos Vip?")

if valor >= 200:
    porcentagemvip = 0.2*valor
    porcentagemcomum = 0.1*valor
    descontovip = valor - porcentagemvip 
    descontocomum = valor - porcentagemcomum 

    if Vip == "sim":
        print(f"Valor descontado é {porcentagemvip}. E o valor a se descontar no final é {descontovip}")
    else:
        print(f"Valor descontado é {porcentagemcomum}. E o valor a se descontar no final é {descontocomum}")
else:
    print("não tem desconto")
  
```
