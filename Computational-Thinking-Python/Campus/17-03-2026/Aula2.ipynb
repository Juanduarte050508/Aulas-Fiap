```python
#10 Equiação 2° Grau

a= 10 
b= 3
c= 1

#Calculo
delta= b*b -4*a*c
bhaskaraX1 = (-b + delta**0.5)/(2 * a)
bhaskaraX2 = (-b - delta**0.5)/(2 * a)

#Saida
print(delta)
print("Valor positivo de Bhaskara", bhaskaraX1)
print("Valor negativo de Bhaskara", bhaskaraX2)
```
